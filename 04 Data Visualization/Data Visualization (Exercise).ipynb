### Dessert Popularity
***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dessert_sales = {
  'Lava Cake': 14,
  'Mousse': 5,
  'Chocolate Cake': 12,
  'Ice Cream': 19,
  'Truffles': 6,
  'Brownie': 8,
  'Chocolate Chip Cookie': 12,
  'Chocolate Pudding': 9,
  'Souffle': 10,
  'Chocolate Cheesecake': 17,
  'Chocolate Chips': 2,
  'Fudge': 9,
  'Mochi': 13,
}

desserts = pd.DataFrame({'dessert': dessert_sales.keys(), 'sales': dessert_sales.values()})
desserts.sort_values('sales', inplace=True)

colors = np.full(len(desserts), 'b')
colors[:3] = 'r'

plt.figure(figsize=(16, 8))
plt.title('Dessert Sales in the Last Week')
plt.ylabel('Sales')
plt.bar(np.arange(0, len(desserts)), desserts.sales, tick_label=desserts.dessert, color=colors)
plt.xticks(rotation=90)
plt.show()
```
